```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("../data/05_song_sentiment_scores.csv")
```

```python
df
```

```python
sum_df = df.groupby(['singer','label']).sum({'score'})
```

```python
sum_df
```

```python
df["new_score"] = df[df['label']=="negative"]['score'].apply(lambda x: x*-1)

        #df['scoree'] = df['score'][i].apply(lambda x: x*-1)
```

```python
df
```

```python
df["new_score2"] = df[df['label']=="positive"]['score'].apply(lambda x: x*1)
```

```python
df
```

```python
df = df.fillna(0)
```

```python
df
```

```python
df["new_score3"] = df["new_score"] + df["new_score2"]
```

```python
df
```

```python
df_new = df[["singer", "lyrics", "new_score3"]]
```

```python
df_new
```

```python
df_new.rename(columns = {"new_score3": "score"}, inplace = True)
```

```python
df_new
```

```python
df_new.to_csv("../data/08_song_sentimentscores_new.csv", index=False)
```

```python
gk = df_new.groupby('singer')['score'].sum()
```

```python
gk
```

```python
df_new.groupby(by="singer").sum().sort_values(by="score").score.plot.bar(color = "purple", title = "total singer rating")
plt.show
```

```python
df_av_score = df_new.groupby(['singer']).mean()
```

```python
df_av_score
```

```python
df_av_score.groupby(by="singer").sum().sort_values(by="score").score.plot.bar(color = "brown", title = "average singer rating")
plt.show
```
