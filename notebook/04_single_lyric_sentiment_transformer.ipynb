```python
import pandas as pd
```

```python
df = pd.read_csv("../data/01_raw_lyrics_dataset.csv") #notebook ile csv aynı dizinde
```

```python
from transformers import AutoModelForSequenceClassification, AutoTokenizer, pipeline

model = AutoModelForSequenceClassification.from_pretrained("savasy/bert-base-turkish-sentiment-cased")
tokenizer = AutoTokenizer.from_pretrained("savasy/bert-base-turkish-sentiment-cased")
sa= pipeline("sentiment-analysis", tokenizer=tokenizer, model=model)

print(sa(df.lyrics[250]))
print("")
print(df.lyrics[250])
```

```python
from sentence_transformers import SentenceTransformer
sentences = ["Bu örnek bir cümle", "Her cümle vektöre çevriliyor", "Sonuncu örnek kötü"]

model = SentenceTransformer('emrecan/bert-base-turkish-cased-mean-nli-stsb-tr')
```

```python
vectors = model.encode(df["lyrics"][250])
```

```python
import numpy as np
```

```python
normalized_vectors = vectors / np.sqrt(np.sum(vectors**2))
```
