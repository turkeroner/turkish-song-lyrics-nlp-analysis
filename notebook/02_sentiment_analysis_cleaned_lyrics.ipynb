```python
import pandas as pd
import numpy as np
import nltk
from nltk.corpus import stopwords
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("turkish_song_lyrics.csv")
```

```python
df
```

### DataFrame düzenleme

```python
df = df.loc[:,["singer", "lyrics"]]
```

```python
df.head()
```

```python
df["No"] = df.index+1
```

```python
df = df.set_index("No")
```

```python
df.head()
```

### cleaning

```python
df.lyrics = df.lyrics.str.replace("\d","")
```

```python
df.head()
```

```python
sw = stopwords.words('turkish')
ex_sw = ["sen", "böyle", "şöyle", "mi", "ben", "yine",'i', "bi", "gene",
         "nakarat", "verse", "x", "kimden", "ne", "sözmüzik", "sana",
         "ona", "bana", "oh", "oy", "ha", "bize", "size","seni","beni",
         "bir","söz", "müzik", "le"]
sw.extend(ex_sw)
print(sw)
```

```python
df["no_stop"] = df["lyrics"].str.lower().apply(lambda x: ' '.join([word for word in x.split() if word not in (sw)]))
```

```python
df.head()
```

```python
df.no_stop[251]
```

### bert-sentiment

```python
from transformers import AutoModelForSequenceClassification, AutoTokenizer, pipeline

model = AutoModelForSequenceClassification.from_pretrained("savasy/bert-base-turkish-sentiment-cased")
tokenizer = AutoTokenizer.from_pretrained("savasy/bert-base-turkish-sentiment-cased")
sa= pipeline("sentiment-analysis", tokenizer=tokenizer, model=model)
```

```python
list_no_stop = df.no_stop.tolist()
```

```python
söz_score = []
for söz in range(len(list_no_stop)):
    score = sa(list_no_stop[söz], truncation=True)
    söz_score.append(score)
```

```python
söz_score[250]
```

```python
list_label = []
for i in range(len(söz_score)):
    for j in range(1):
        list_label.append(söz_score[i][j]["label"])
```

```python
list_label[250]
```

```python
list_score = []
for i in range(len(söz_score)):
    for j in range(1):
        list_score.append(söz_score[i][j]["score"])
```

```python
list_score[250]
```

```python
df["+/-"] = list_label
```

```python
df["Score"] = np.array(list_score)
```
